## Loading a dataset

```python
import pandas as pd
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'label'])
iris

# 길이가 길면 너비도 크고... correlation이 있으니까 차원을 축소해보자는 것.
```

```python
X = iris.drop(['label'], axis=1)
X
```

## Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)  # n_components specifies the new dimensionality 여기에 5가 들어가면 안 된대 왜?
# 네 개의 차원을 가지고 분석을 했는데 다섯 개로 분석해달라고 하는 건 말이 안됨
# 이것이 2차원을 분석해달라고 요구한 줄임.
X_reduced = model.fit(X).transform(X) # or just pca.fit_transform(X)
pd.DataFrame(X_reduced)

# 데이터가 길게 퍼져있는 쪽이 pc1, 아닌 쪽이 pc2. 그 축을 기준으로 읽어준 값이래...
```

```python
print('Components:', model.components_)
# 축의 방향 - 원점으로 평행이동해서 (,) 으로 나타내면 방향을 나타낼 수 있음(이차원에서) - 4차원에서는 숫자 네 개로.
# component의 4개는 pc1, pc2의 방향을 나타낸다고 생각하면 됨(정확히는 벡터.)

print('Explained Variance:', model.explained_variance_)
# 데이터의 폭에 해당되는 값. 원반의 두께. pc1이 당연히 크고 pc2가 작음. (큰 값, 작은 값, 더 작은 값 ...)
print('Explained Variance Ratio:', model.explained_variance_ratio_ )
# 총 네 개의 숫자로 이 데이터 전체를 설명했는데 이를 두 개로 줄였음. 그랬더니 조금의 오차가 있을 수 있음. 얼만큼 부정확한가를 알려줌.
# 정확도를 알려줌. 꽃받침과 꽃잎조차도 correaltion 있기 때문에 숫자 하나로 해도 되겠다... 같은 느낌. 92%나 나오기 때문에.(숫자 하나만 가지고도 92%가 설명 가능하다)
# 하나 더 쓰면 97.5%까지 커버 가능

```

```python
X_approx = model.inverse_transform(X_reduced) # inverse - 좀 부정확해도 숫자 두 개를 가지고 다시 4개로 복원 가능. 2%만 잃었으니까. 그 부정확한 정도를 보자
pd.DataFrame(X_approx)

# 이차원으로 만들어진 값들에 윗줄 component: [[~]] 곱하면 나옴. 하... 모르면 안 내면 안 되실까요..
# 행렬은 직사각형으로 생긴 숫자. 1x2 = 1행 2열. [ㅇㅇ] / 2x1 => (2)둘이 같으니까 곱하기가 됨. 둘이 같을 때만 곱하기가 됨.
# 좆같네... [ㅇㅇ] x [ㅁ/ㅅ] -> ㅁㅅ 형태로 돌려 곱하고 둘이 더함. 위의 것이 먼저 와야 함. 만약 뒤의 것이 [ㅁㅅ/ㅎㅍ]이면 ㅁㅎ, ㅅㅍ형태로 바꾸어서 곱해주ㅓ야 함.
# pca한 결과값 - 150x2행렬임. 그리고 componenets는 2x4행렬. 2가 동일!! 그러면 곱하면 결과값이 총 네 개의 값이 150개. 150x4의 행렬이 됨.
# components의 두 벡터는 반드시 직각. 사차원에서 점을 찍어 원점으로부터 화살표를 그리면 그 두 개는 직각을 이룸. 왜 자꾸 글자 잡아먹음 개자식이

# 첫 번째 꽃받침 5.08 - 원래 5.1이었음. 비슷한 것을 볼 수 있음.
# 이걸 왜 하는 걸까? 두 가지 이유가 있는데 사회과학에서 인과관계를 쓰려면 필수. 두 번째는 학문적으로 의미를 줄 수 있음. correlation은 별로 좋지 않다. decorrelate된 독립적인 정보를 얻음.
# 애초에 축을 두는 게 decorrelate하려고 한 건데 놓고 보니 또 correlate해서 다시 바꾸는 것임.
# 계산량을 줄이기 위해서. 네 개를 모두 계산하는 건 낭비니까.
```
